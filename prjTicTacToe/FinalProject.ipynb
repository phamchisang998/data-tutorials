```python
from functools import *
def coordinate_update(next_player, coordinate):
    Flag = True
    while(Flag):
        coor = input()
        if not coor.isnumeric() :
            Flag = True
            continue
        if int(coor) not in coordinate.keys():
            Flag = True
            continue
        if coordinate.get(int(coor)) != " ":
            Flag = True
            print("Invalid coordinate, please select a other one: ")
            continue
        Flag = False
    coordinate[int(coor)] = next_player
    return {k: v for k, v in coordinate.items() if v != " "}
```
