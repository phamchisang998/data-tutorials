```python
import os
import random
import json

class GenerateData:
    def __init__(self, locationPath, schoolName):
        self.locationPath = locationPath
        self.schoolName = schoolName
        self.schoolPath = self.locationPath + self.schoolName + "/"
        print(self.schoolPath)
        self.grade10Path = self.schoolPath + "Grade10/"
        print(self.grade10Path)
        self.grade11Path = self.schoolPath + "Grade11/"
        print(self.grade11Path)
        self.grade12Path = self.schoolPath + "Grade12/"
        print(self.grade12Path)
        
    def createStorage(self):
        schDir = self.schoolName
        parent_schDir = self.locationPath
        schPath = os.path.join(parent_schDir, schDir)
        os.mkdir(schPath)

        grade10Path = os.path.join(self.schoolPath, "Grade10")
        os.mkdir(grade10Path)
        grade10Path = os.path.join(self.schoolPath, "Grade11")
        os.mkdir(grade10Path)
        grade10Path = os.path.join(self.schoolPath, "Grade12")
        os.mkdir(grade10Path)

    def generateStudentData(self, numGrade10, numGrade11, numGrade12):
        count_schID = 0
        for i in range(numGrade10):
            classPath = self.grade10Path + "10A" + str(i + 1) + ".json"
            print(classPath)
            classInfo = []

            for stu in range(random.randint(30, 36)):
                count_schID = count_schID + 1
                count_classID = stu + 1

                scores = {}
                scores["Toan"] = round(random.uniform(0.00,10.00), 2)
                scores["Li"]   = round(random.uniform(0.00,10.00), 2)
                scores["Hoa"]  = round(random.uniform(0.00,10.00), 2)
                scores["Van"]  = round(random.uniform(0.00,10.00), 2)
                scores["Su"]   = round(random.uniform(0.00,10.00), 2)
                scores["Dia"]  = round(random.uniform(0.00,10.00), 2)
                scores["Anh"]  = round(random.uniform(0.00,10.00), 2)
                scores["CN"]   = round(random.uniform(0.00,10.00), 2)
                scores["CD"]   = round(random.uniform(0.00,10.00), 2)
                scores["Tin"]  = round(random.uniform(0.00,10.00), 2)
                scores["Sinh"] = round(random.uniform(0.00,10.00), 2)
                
                stuInfo = {"schoolID": count_schID, "classID": count_classID, "scores": scores}
                classInfo.append(stuInfo)

            jsonObject = json.dumps(classInfo)
            with open(classPath, "w") as outfile:
                    outfile.write(jsonObject)

        for i in range(numGrade11):
            classPath = self.grade11Path + "11B" + str(i + 1) + ".json"
            print(classPath)
            classInfo = []

            for stu in range(random.randint(30, 36)):
                count_schID = count_schID + 1
                count_classID = stu + 1

                scores = {}
                scores["Toan"] = round(random.uniform(0.00,10.00), 2)
                scores["Li"]   = round(random.uniform(0.00,10.00), 2)
                scores["Hoa"]  = round(random.uniform(0.00,10.00), 2)
                scores["Van"]  = round(random.uniform(0.00,10.00), 2)
                scores["Su"]   = round(random.uniform(0.00,10.00), 2)
                scores["Dia"]  = round(random.uniform(0.00,10.00), 2)
                scores["Anh"]  = round(random.uniform(0.00,10.00), 2)
                scores["CN"]   = round(random.uniform(0.00,10.00), 2)
                scores["CD"]   = round(random.uniform(0.00,10.00), 2)
                scores["Tin"]  = round(random.uniform(0.00,10.00), 2)
                scores["Sinh"] = round(random.uniform(0.00,10.00), 2)
                
                stuInfo = {"schoolID": count_schID, "classID": count_classID, "scores": scores}
                classInfo.append(stuInfo)

            jsonObject = json.dumps(classInfo)
            with open(classPath, "w") as outfile:
                    outfile.write(jsonObject)

        for i in range(numGrade12):
            classPath = self.grade12Path + "12C" + str(i + 1) + ".json"
            print(classPath)
            classInfo = []

            for stu in range(random.randint(30, 36)):
                count_schID = count_schID + 1
                count_classID = stu + 1

                scores = {}
                scores["Toan"] = round(random.uniform(0.00,10.00), 2)
                scores["Li"]   = round(random.uniform(0.00,10.00), 2)
                scores["Hoa"]  = round(random.uniform(0.00,10.00), 2)
                scores["Van"]  = round(random.uniform(0.00,10.00), 2)
                scores["Su"]   = round(random.uniform(0.00,10.00), 2)
                scores["Dia"]  = round(random.uniform(0.00,10.00), 2)
                scores["Anh"]  = round(random.uniform(0.00,10.00), 2)
                scores["CN"]   = round(random.uniform(0.00,10.00), 2)
                scores["CD"]   = round(random.uniform(0.00,10.00), 2)
                scores["Tin"]  = round(random.uniform(0.00,10.00), 2)
                scores["Sinh"] = round(random.uniform(0.00,10.00), 2)
                
                stuInfo = {"schoolID": count_schID, "classID": count_classID, "scores": scores}
                classInfo.append(stuInfo)

            jsonObject = json.dumps(classInfo)
            with open(classPath, "w") as outfile:
                    outfile.write(jsonObject)

    def getCurrentPath():
        # check the current path to see where we are
        # then add the template files
        pass

    def createFile(self):
        pass

    def rankingStudent(self, numGrade10, numGrade11, numGrade12):
        for i in range(numGrade10):
            classPath = self.grade10Path + "10A" + str(i + 1) + ".json"
            resultClassPath = self.grade10Path + "10A" + str(i + 1) + "_ranked" + ".json"
            jsonFile = open(classPath)
            classInfo = json.load(jsonFile)

            rankedClassInfo = list(map(self.addRanking, classInfo))
            # print(rankedClassInfo)
            jsonObject = json.dumps(rankedClassInfo)
            with open(resultClassPath, "w") as outfile:
                            outfile.write(jsonObject)

        for i in range(numGrade11):
            classPath = self.grade11Path + "11B" + str(i + 1) + ".json"
            resultClassPath = self.grade11Path + "11B" + str(i + 1) + "_ranked" + ".json"
            jsonFile = open(classPath)
            classInfo = json.load(jsonFile)

            rankedClassInfo = list(map(self.addRanking, classInfo))
            # print(rankedClassInfo)
            jsonObject = json.dumps(rankedClassInfo)
            with open(resultClassPath, "w") as outfile:
                            outfile.write(jsonObject)

        for i in range(numGrade12):
            classPath = self.grade12Path + "12C" + str(i + 1) + ".json"
            resultClassPath = self.grade12Path + "12C" + str(i + 1) + "_ranked" + ".json"
            jsonFile = open(classPath)
            classInfo = json.load(jsonFile)

            rankedClassInfo = list(map(self.addRanking, classInfo))
            # print(rankedClassInfo)
            jsonObject = json.dumps(rankedClassInfo)
            with open(resultClassPath, "w") as outfile:
                            outfile.write(jsonObject)
    
    def addRanking(self, studentInfo):
        studentInfo["rank"] = self.ranking(studentInfo["scores"])
        return studentInfo

    def ranking(self, scores):
        avgScore = float(0)
        for k,v in scores.items():
            avgScore = avgScore + v
        avgScore = avgScore / 11

        allGreater6p5 = True
        for k,v in scores.items():
            allGreater6p5 = allGreater6p5 and (v >= 6.5)
        # print("allGreater6p5", allGreater6p5)
        allGreater5p0 = True
        for k,v in scores.items():
            allGreater5p0 = allGreater5p0 and (v >= 5.0)
        # print("allGreater5p0", allGreater5p0)
        allGreater3p5 = True
        for k,v in scores.items():
            allGreater3p5 = allGreater3p5 and (v >= 3.5)
        # print("allGreater3p5", allGreater3p5)


        if avgScore >= 8.0 and allGreater6p5 and (scores["Toan"] >= 8.0 or scores["Van"] >= 8.0):
            return "Gioi"
        elif avgScore >= 6.5 and allGreater5p0 and (scores["Toan"] >= 6.5 or scores["Van"] >= 6.5):
            return "Kha"
        elif avgScore >= 5.0 and allGreater3p5 and (scores["Toan"] >= 5.0 or scores["Van"] >= 5.0):
            return "TB"
        else:
            return "Yeu"
        
    def sumStuType(self, numGrade10, numGrade11, numGrade12):
        # use reduce() mmethod
        for i in range(numGrade10):
            # classPath = self.grade10Path + "10A" + str(i + 1) + ".json"
            resultClassPath = self.grade10Path + "10A" + str(i + 1) + "_ranked" + ".json"
            jsonFile = open(resultClassPath)
            classInfo = json.load(jsonFile)
            from functools import reduce
            sumInfo = reduce(
                self.increase_ranks
                ,classInfo 
            )

            print(sumInfo)
        
    def increase_ranks(self, x, y):
        y[x.get('rank')] = y.get(x.get('rank'), 0) + 1
        return y
```
